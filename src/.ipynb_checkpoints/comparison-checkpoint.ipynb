```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
from math import pi

# Set data
df = pd.DataFrame({
'group':            ['NW', 'Hirschberg\'75', 'Dijkstra\'85', 'biWFA', 'Edlib', 'A*PA'],
'similar long':    [ 1,   3,                5,              7,        7,       3],
'fast for different': [ 10,  3,                5,              7,        7,       3],
'efficient':        [ 1,   3,                5,              7,        7,       3],
    
#'cells/sec':        [ 10,  7,                5,              7,        7,       3],
#'group':            ['NW\'70+Gotoh\'82', 'Hirschberg\'75', 'Dijkstra\'85', 'Diagonal transition\'85', 'Exp.band\'85', 'A*\'22'],
#'scales time':      [ 1,                  1,                1,              5,                         5,              10],
#'scales memory':    [ 1,                  10,               1,              10,                        5,              10],
    #'affine':           [ 10,                 3,                3,              8,                         1,              1]
})

 
# ------- PART 1: Create background
 
# number of variable
categories=list(df)[1:]
N = len(categories)

# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels
plt.xticks(angles[:-1], categories)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([2,4,6,8], ["2","4","6","8"], color="grey", size=7)
plt.ylim(0,10)

ax.set_yticks([])
#plt.axis('off')
#print(list(ax.spines))
ax.spines[['polar', 'start', 'end', 'inner']].set_visible(False)

# ------- PART 2: Add plots
 
# Plot each individual = each line of the data
# I don't make a loop, because plotting more than 3 groups makes the chart unreadable

which  = [1, 0, 0, 1, 1, 1]
colors = ['blue', 'red', 'yellow', 'brown', 'pink', 'black']

for i in range(6):
    if which[i]:
        values=df.loc[i].drop('group').values.flatten().tolist()
        values += values[:1]
        ax.plot(angles, values, linewidth=1, linestyle='solid', label=df.group[i])
        ax.fill(angles, values, colors[i], alpha=0.1)
 
# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show the graph
plt.show()
```
